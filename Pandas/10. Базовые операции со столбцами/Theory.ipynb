```python
import pandas as pd
```

```python
melb_data = pd.read_csv('data/melb_data_ps.csv', sep=',')
melb_data.head()
```

## Создание копии таблицы

```python
melb_df = melb_data.copy()
melb_df.head()
```

## Удаление столбцов
Мы хотим построить модель, которая бы предсказывала цену объекта недвижимости в Мельбурне на основе выводов:
- Цена объекта никак не зависит от его порядкового номера (столбец index)
- Признак, описывающий широту и долготу в виде кортежа Coordinates, дублирует информацию, представленную в столбцах Longitude & Lattitude
## За удаление строк и столбцов в таблице отвечает метод .drop()
## Основные параметры методо .drop()
- labels - порядковые номера или имена столбцов, которые подлежат удалению, если их несколько, они передаются в список
- axis - ось совершения операции, axis=0 - удаляются строки, axis=1 - удаляются столбцы
- inplace - если параметр выставлен на True, происходит замена изначального DataFrame на новый, при этом метод ничего не возвращает, если на False - возвращается копия DataFrame, из которой удалены указанные строки (столбцы), при этом изначальный DataFrame не изменяется, дефолтный параметр = False

```python
# Удалим столбцы index и coordinates из таблицы, выведем первые 5 строк
melb_df = melb_df.drop(['index', 'Coordinates'], axis=1)
melb_df.head()
```

## Математические операции со столбцами
В DataFrame можно производить мат. операции со столбцами, причем они будут гораздо быстрее операций с циклами, достигается это за счет того, что они выполняются на языке программирования С

```python
# Например создадим переменную total_rooms, в которой будем хранить общее количество комнат в здании
total_rooms = melb_df['Rooms'] + melb_df['Bathroom'] + melb_df['Bedroom']
display(total_rooms)
```

```python
# теперь введем признак MeanRoomsArea, который соответствует средней площади одной комнаты для каждого объекта.
# Для этого разделим площадь здания на полученное ранее количество комнат
melb_df['MeanRoomsArea'] = melb_df['BuildingArea'] / total_rooms
display(melb_df['MeanRoomsArea'])
```

```python
# Можно ввести еще один признак AreaRatio, коэффициент соотношения площади здания (BuildingArea) и площади участка (Landlize)
# Для этого разницу двух площадей поделим на их сумму
diff_area = melb_df['BuildingArea'] - melb_df['Landsize']
sum_area = melb_df['BuildingArea'] + melb_df['Landsize']
melb_df['AreaRatio'] = diff_area / sum_area
display(melb_df['AreaRatio'])
```
